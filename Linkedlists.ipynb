```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

class LinkedList:
    def __init__(self):
        self.head = None

    def append(self, name, adm_no, grades):


        node_data = {
            'Name': name,
            'ADM NO': adm_no,
            'Grades': grades
        }
        new_node = Node(node_data)

        if self.head is None:
            self.head = new_node
            return

        last_node = self.head
        while last_node.next:
            last_node = last_node.next
        last_node.next = new_node

    def display(self):
        current = self.head
        if not current:
            print("The linked list is empty.")
            return

        print("\n Linked List Content ")
        node_num = 1
        while current:
            print(f"Node {node_num}:")
            for key, value in current.data.items():
                if key == 'Grades':
                    print(f"  {key}:")
                    for grade_entry in value:
                        print(f"    Unit: {grade_entry['Unit']}, Grade: {grade_entry['Grade']}")
                else:
                    print(f"  {key}: {value}")
            current = current.next
            node_num += 1


```

```python
my_linked_list = LinkedList()


my_linked_list.append(
    name='Alice Smith',
    adm_no='ADM001',
    grades=[
        {'Unit': 'Mathematics', 'Grade': 'A'},
        {'Unit': 'Physics', 'Grade': 'B+'},
        {'Unit': 'Chemistry', 'Grade': 'A-'}
    ]
)


my_linked_list.append(
    name='Bob Johnson',
    adm_no='ADM002',
    grades=[
        {'Unit': 'Mathematics', 'Grade': 'B'},
        {'Unit': 'Physics', 'Grade': 'C+'},
        {'Unit': 'Biology', 'Grade': 'B-'}
    ]
)


my_linked_list.append(
    name='Charlie Brown',
    adm_no='ADM003',
    grades=[
        {'Unit': 'Literature', 'Grade': 'A'},
        {'Unit': 'History', 'Grade': 'A'},
        {'Unit': 'Geography', 'Grade': 'B+'}
    ]
)


my_linked_list.display()
```
