```python
import numpy as np

# Task 1: Create a 2D array (4 branches × 6 months)
transactions = np.array([
    [1200, 1350, 1100, 1450, 1600, 1700],  
    [1000, 1250, 1300, 1400, 1500, 1600],  
    [900, 950, 1000, 1050, 1100, 1150],    
    [1500, 1550, 1600, 1650, 1700, 1750]   
])

# Task 2: Total transactions per branch
total_per_branch = np.sum(transactions, axis=1)

# Task 3: Branch with highest total transactions
highest_branch_index = np.argmax(total_per_branch)
highest_branch_total = total_per_branch[highest_branch_index]

# Task 4: Average monthly transaction volume across all branches
average_monthly_volume = np.mean(transactions)

# Task 5: Reshape to 3x8 and explain
reshaped = transactions.reshape(3, 8)
implication = "Reshaping changes the structure but not the data. It may lose the original branch/month meaning."
```

QUESTION 2

```python
# Task 1: Create list of 10 delivery routes
routes = [
    "Nairobi-Kisumu", "Mombasa-Nairobi", "Eldoret-Nakuru", "Kisumu-Busia",
    "Nairobi-Eldoret", "Nakuru-Mombasa", "Busia-Kakamega", "Kakamega-Kisii",
    "Nairobi-Thika", "Thika-Garissa"
]
print(routes)
```

```python
# Task 2: Append new route and remove discontinued one
routes.append("Garissa-Wajir")
print(routes)
```

```python
routes.remove("Kisumu-Busia")
print(routes)
```

```python
# Task 3: Sort alphabetically and reverse
routes.sort()
print(routes)
```

```python
routes.reverse()
print(routes)
```

```python
# Task 4: Count routes starting with "N"
count_N = sum(route.startswith("N") for route in routes)
print(count_N)
```

```python
# Task 5: List comprehension for routes longer than 10 characters
long_routes = [route for route in routes if len(route) > 10]
print(long_routes)
```
